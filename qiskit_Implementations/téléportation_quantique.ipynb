```python
#teleportation 
from qiskit import*
circuit1 = QuantumCircuit(3,2)

#intrication de bob et alice 
circuit1.h(1)
circuit1.cx(1,2)

circuit1.draw('mpl')
```

```python
#simulation
from qiskit_aer import Aer
from qiskit.visualization import plot_histogram
```

```python
#intrication de avec l'autre
circuit1.cx(0,1)
circuit1.h(0)

circuit1.measure(0,0)
circuit1.measure(1,1)
circuit1.x(2).c_if(1,1)  # Si le deuxième bit mesuré est 1, appliquer une porte X
circuit1.z(2).c_if(0,1)  # Si le premier bit mesuré est 1, appliquer une porte Z
circuit1.draw('mpl')

```

```python
circuit1.measure_all()
Aer.backends()
simulator = Aer.get_backend('aer_simulator')
compiled_circuit = transpile(circuit1, simulator)
job = simulator.run(compiled_circuit, shots=1000)
result = job.result()
counts = result.get_counts(circuit1)
plot_histogram(counts, title = "Histogramme")
```
